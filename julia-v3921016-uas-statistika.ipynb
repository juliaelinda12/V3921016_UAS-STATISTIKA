```python
df = pd.read_csv("../input/julia-uas-statistika/Jan_2019_ontime.csv")
df.shape
```

```python
#menghapus kolom yang tidak kitagunakan
df = df.drop(columns=['DAY_OF_MONTH','DAY_OF_WEEK','OP_UNIQUE_CARRIER',
                           'OP_CARRIER_AIRLINE_ID','OP_CARRIER','TAIL_NUM',
                           'OP_CARRIER_FL_NUM','ORIGIN_AIRPORT_ID','DEST_AIRPORT_ID',
                           'DEP_TIME_BLK','ARR_TIME','CANCELLED','DIVERTED','DISTANCE'])
data = df.head(350)
data.head()
```

```python
#seleksi data X dan Y
X = data.iloc[:, :4].values
Y = data.iloc[:, 4].values
```

```python
#mengubah label data categorical ini menjadi numerik
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:, 0] = le.fit_transform(X[:, 0])
X[:, 1] = le.fit_transform(X[:, 1])
```

```python
#mencetak data yang sudah diubah
print(X)
```

```python
#memisahkan data training dan data set
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.2, random_state = 0)

print(x_test)
```

```python
#Feature scaling
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)

print(x_train)
```

```python
#Pembuatan Model training
from sklearn.naive_bayes import GaussianNB

classifier = GaussianNB()
classifier.fit(x_train, y_train)
```

```python
#Memprediksi hasil dari dataset
y_pred = classifier.predict(x_test)

print(y_pred)
```

```python
#Cek akurasi dari model training
from sklearn.metrics import accuracy_score

print('Model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

```python
#Cek nilai akurasi
from sklearn.metrics import confusion_matrix,accuracy_score

ac = accuracy_score(y_test, y_pred)

ac
```

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

print('Confusion matrix\n\n', cm)

print('\nTrue Positives(TP) = ', cm[0,0])

print('\nTrue Negatives(TN) = ', cm[1,1])

print('\nFalse Positives(FP) = ', cm[0,1])

print('\nFalse Negatives(FN) = ', cm[1,0])
```

```python
#Classification matrix
cm_matrix = pd.DataFrame(data=cm, columns=['Actual Positive:1', 'Actual Negative:0'], 
                                 index=['Predict Positive:1', 'Predict Negative:0'])

sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='YlGnBu')
```

```python
TP = cm[0,0]
TN = cm[1,1]
FP = cm[0,1]
FN = cm[1,0]
```

```python
# print classification accuracy

classification_accuracy = (TP + TN) / float(TP + TN + FP + FN)

print('Classification accuracy : {0:0.4f}'.format(classification_accuracy))
```

```python
# print classification error

classification_error = (FP + FN) / float(TP + TN + FP + FN)

print('Classification error : {0:0.4f}'.format(classification_error))
```

```python
precision = TP / float(TP + FP)


print('Precision : {0:0.4f}'.format(precision))
```

```python
recall = TP / float(TP + FN)

print('Recall or Sensitivity : {0:0.4f}'.format(recall))
```

```python
true_positive_rate = TP / float(TP + FN)


print('True Positive Rate : {0:0.4f}'.format(true_positive_rate))
```

```python
false_positive_rate = FP / float(FP + TN)


print('False Positive Rate : {0:0.4f}'.format(false_positive_rate))
```

```python
specificity = TN / (TN + FP)

print('Specificity : {0:0.4f}'.format(specificity))
```
